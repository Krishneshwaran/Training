### To check the csv file in which encoding

### Changing the CSV File encoding using pandas

```python
import pandas as pd

path =  'level1.csv'

df = pd.read_csv(path, encoding='ISO-8859-1')
df.to_csv(path, encoding='utf-8', index=False)
```
