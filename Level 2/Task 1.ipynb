#  Task 1: Simple Plot

```python
import matplotlib.pyplot as plt


x_values=[1,2,3]

y_values=[4,5,6]


plt.plot(x_values, y_values, marker='o', linestyle='-', color='b',  label="y = 2x")


plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Simple Plot')

plt.legend
plt.grid(True)
plt.show()
```
